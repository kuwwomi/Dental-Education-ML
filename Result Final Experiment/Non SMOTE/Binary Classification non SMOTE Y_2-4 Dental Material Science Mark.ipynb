```python
# Kolom yang akan digunakan untuk klasifikasi
grading_columns = ['PATHOLOGY MARK', 'MICROBIOLOGY MARK', 'PHARMACOLOGY MARK', 'DENTAL MATERIAL SCIENCE MARK ']

# Fungsi untuk mengklasifikasikan nilai menjadi kategori
def classify_grades(score):
    # Mengonversi nilai ke tipe data numerik jika belum
    try:
        score = float(score)
    except ValueError:
        return 'Invalid Score'  # Atau tindakan yang sesuai jika nilai tidak dapat dikonversi

    if score >= 75:
        return 'A'
    elif 60 <= score < 75:
        return 'B'
    elif 50 <= score < 60:
        return 'C'
    else:
        return 'F'

# Iterasi melalui kolom-kolom yang akan dijadikan dasar klasifikasi
for column in grading_columns:
    # Membuat kolom baru dengan nama yang dihasilkan dari kolom asli ditambah '_GRADE'
    new_column_name = column + '_GRADE'
    df[new_column_name] = df[column].apply(classify_grades)

# Tampilkan DataFrame yang sudah dimodifikasi
print(df)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

x = df.drop(["STUDENT'S IDENTITY NO. ", 'ANATOMY STATUS', 'PHYSIOLOGY STATUS', 'BIOCHEMISTRY STATUS',
             'ORAL BIOLOGY STATUS', 'PATHOLOGY MARK', 'PATHOLOGY STATUS', 'Batch', 'REPEAT PAPER',
             'MICROBIOLOGY MARK', 'MICROBIOLOGY STATUS', 'PHARMACOLOGY MARK',
             'PHARMACOLOGY STATUS', 'DENTAL MATERIAL SCIENCE MARK ',
             'DENTAL MATERIAL SCIENCE STATUS', 'BIRTH DATE','MIDDLE IC NO.',
             'DISTRICT','POSCODE', 'STATE', 'COUNTRY', 'PRE-UNIVERSITY','LAST DIGIT MATRIC NO.', 'GENDER_MALE', 'PATHOLOGY MARK_GRADE',
             'MICROBIOLOGY MARK_GRADE', 'PHARMACOLOGY MARK_GRADE', 'DENTAL MATERIAL SCIENCE MARK _GRADE'], axis=1)
y = df['DENTAL MATERIAL SCIENCE MARK _GRADE']

k_best = 4  # Ganti dengan jumlah fitur terbaik yang ingin Anda pilih
selector = SelectKBest(score_func=f_classif, k=k_best)
x_new = selector.fit_transform(x, y)

# Menampilkan fitur terpilih
selected_features = x.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
# Memisahkan data menjadi data training dan data testing menjadi 60% data training, 40% data testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = 0.4,
                                                   random_state = 2022)
```

# Modelling

## Logistic Regression

```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix

# Model Development
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=2022)
log_reg.fit(x_train, y_train)

y_pred = log_reg.predict(x_test)
y_pred

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Definisi model Logistic Regression
log_reg = LogisticRegression(random_state=2022)

# Tentukan daftar hyperparameter yang ingin diuji
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

# Inisialisasi GridSearchCV dengan model, hyperparameter grid, dan jumlah lipatan cross-validation
grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='accuracy')

# Latih model dengan melakukan pencarian grid
grid_search.fit(x_train, y_train)

# Cetak hyperparameter terbaik yang ditemukan
print("Best hyperparameters:", grid_search.best_params_)

# Gunakan model dengan hyperparameter terbaik untuk membuat prediksi pada data testing
y_pred_log_tuned = grid_search.best_estimator_.predict(x_test)

# Cetak laporan klasifikasi dan matriks kebingungan
print("\nClassification Report:")
print(classification_report(y_test, y_pred_log_tuned))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_log_tuned))
```

## Neural Networks

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Standarisasi data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Menggunakan LabelEncoder untuk mengonversi label menjadi format numerik
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Membangun model neural network
model = Sequential()
model.add(Dense(64, input_dim=x_train_scaled.shape[1], activation='relu'))  # Layer input
model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dense(1, activation='sigmoid'))  # Layer output, output biner sehingga menggunakan sigmoid

# Kompilasi model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Melatih model
model.fit(x_train_scaled, y_train, epochs=10, batch_size=32, validation_data=(x_test_scaled, y_test))

# Evaluasi model
eval_result = model.evaluate(x_test_scaled, y_test)
print(f'\nTest Loss: {eval_result[0]}, Test Accuracy: {eval_result[1]}')

# Prediksi
y_pred = (model.predict(x_test_scaled) > 0.5).astype("int32")

# Menampilkan hasil evaluasi model
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
