```python
# Kolom yang akan digunakan untuk klasifikasi
grading_columns = ['PATHOLOGY MARK', 'MICROBIOLOGY MARK', 'PHARMACOLOGY MARK', 'DENTAL MATERIAL SCIENCE MARK ']

# Fungsi untuk mengklasifikasikan nilai menjadi kategori
def classify_grades(score):
    # Mengonversi nilai ke tipe data numerik jika belum
    try:
        score = float(score)
    except ValueError:
        return 'Invalid Score'  # Atau tindakan yang sesuai jika nilai tidak dapat dikonversi

    if score >= 75:
        return 'A'
    elif 60 <= score < 75:
        return 'B'
    elif 50 <= score < 60:
        return 'C'
    else:
        return 'F'

# Iterasi melalui kolom-kolom yang akan dijadikan dasar klasifikasi
for column in grading_columns:
    # Membuat kolom baru dengan nama yang dihasilkan dari kolom asli ditambah '_GRADE'
    new_column_name = column + '_GRADE'
    df[new_column_name] = df[column].apply(classify_grades)

# Tampilkan DataFrame yang sudah dimodifikasi
print(df)
```

```python
# Fungsi untuk mengubah nilai ke float dengan menangani kasus '79>80'
def convert_to_float(value):
    try:
        # Jika nilai dapat diubah menjadi float, kembalikan float dari nilai tersebut
        return float(value)
    except ValueError:
        # Jika terjadi ValueError, berarti nilai memiliki format seperti '79>80'
        # Ambil nilai pertama dari format tersebut (sampai karakter '>') dan kembalikan float-nya
        return float(value.split('>')[0])

# Terapkan fungsi convert_to_float pada kolom 'PHYSIOLOGY MARK'
df['PHYSIOLOGY MARK'] = df['PHYSIOLOGY MARK'].apply(convert_to_float)

# Terapkan fungsi convert_to_float pada kolom 'DENTAL MATERIAL SCIENCE MARK'
df['DENTAL MATERIAL SCIENCE MARK '] = df['DENTAL MATERIAL SCIENCE MARK '].apply(convert_to_float)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

x = df.drop(["STUDENT'S IDENTITY NO. ", 'ANATOMY STATUS', 'PHYSIOLOGY STATUS', 'BIOCHEMISTRY STATUS',
             'ORAL BIOLOGY STATUS', 'PATHOLOGY MARK', 'PATHOLOGY STATUS', 'Batch', 'REPEAT PAPER',
             'MICROBIOLOGY MARK', 'MICROBIOLOGY STATUS', 'PHARMACOLOGY MARK',
             'PHARMACOLOGY STATUS', 'DENTAL MATERIAL SCIENCE MARK ',
             'DENTAL MATERIAL SCIENCE STATUS', 'BIRTH DATE','MIDDLE IC NO.',
             'DISTRICT','POSCODE', 'STATE', 'COUNTRY', 'PRE-UNIVERSITY','LAST DIGIT MATRIC NO.', 'GENDER_MALE', 'PATHOLOGY MARK_GRADE',
             'MICROBIOLOGY MARK_GRADE', 'PHARMACOLOGY MARK_GRADE', 'DENTAL MATERIAL SCIENCE MARK _GRADE'], axis=1)
y = df['MICROBIOLOGY MARK_GRADE']

k_best = 4  # Ganti dengan jumlah fitur terbaik yang ingin Anda pilih
selector = SelectKBest(score_func=f_classif, k=k_best)
x_new = selector.fit_transform(x, y)

# Menampilkan fitur terpilih
selected_features = x.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Definisikan model Naive Bayes
naive_bayes_model = GaussianNB()

# Parameter yang ingin dioptimalkan (hanya var_smoothing pada Gaussian Naive Bayes)
param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}

# Gunakan GridSearchCV untuk mencari kombinasi hyperparameter terbaik
grid_search = GridSearchCV(naive_bayes_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# Ambil model dengan kombinasi hyperparameter terbaik
best_naive_bayes_model = grid_search.best_estimator_

# Latih model dengan data pelatihan
best_naive_bayes_model.fit(x_train, y_train)

# Prediksi menggunakan model yang dioptimalkan
y_pred = best_naive_bayes_model.predict(x_test)

# Tampilkan hasil evaluasi
print("Best Parameters:", grid_search.best_params_)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

###NEURAL NETWORK

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Standarisasi data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Menggunakan LabelEncoder untuk mengonversi label menjadi format numerik
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Membangun model neural network
model = Sequential()
model.add(Dense(64, input_dim=x_train_scaled.shape[1], activation='relu'))  # Layer input
model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dense(1, activation='sigmoid'))  # Layer output, output biner sehingga menggunakan sigmoid

# Kompilasi model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Melatih model
model.fit(x_train_scaled, y_train, epochs=10, batch_size=32, validation_data=(x_test_scaled, y_test))

# Evaluasi model
eval_result = model.evaluate(x_test_scaled, y_test)
print(f'\nTest Loss: {eval_result[0]}, Test Accuracy: {eval_result[1]}')

# Prediksi
y_pred = (model.predict(x_test_scaled) > 0.5).astype("int32")

# Menampilkan hasil evaluasi model
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
